```python
import os #basic foldering functions
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip #this is the video clipping function
import pandas as pd #used here for data reading (csv)
import numpy as np #for saving data in array format
from moviepy.video.io.VideoFileClip import VideoFileClip #alternative
#foldering
annotations = pd.read_csv('../motifs/all_groups.csv') #folder of the annotations
name_conversions = pd.read_csv('./info_name_mapping_LP.csv')
vidfolder = '../video_customanimations/'

#what video file .mov .mp4 .avi
vidformat = ".mp4"

outputfol = "../videosanimated_snipped_by_motifs/"

name_conversions.head()
```

```python
import moviepy.editor

def extract_subclip(video, begintime, endtime, targetname):
    clip = VideoFileClip(video).subclip(begintime, endtime)
    #clip_resized = clip.resize(height=480)
    #clip_resized = clip.resize(width=720)
    #clip_resized = clip.resize(1)
    clip.write_videofile(targetname)

for index, row in annotations.iterrows(): #loop through your annoation files
    ID = row['track']   #the first code will be ID that you can match with your video
    ID = name_conversions.loc[name_conversions['Audio file'] == ID, 'display_name'].values[0]
    video = vidfolder+ID+vidformat #get the video that has a matching code
    print(video)
    if os.path.exists(video) == True:
        begintime = row['start'] #extract begintime
        endtime = row['end'] #extract end time
        annotlabel = row['index'] #extract your annotation
        #now take the video and select the relevant chunk and write to the outputfolder
        #ffmpeg_extract_subclip(video, begintime, endtime, targetname=outputfol+str(annotlabel)+vidformat)
        #alternative
        targetname =outputfol+str(annotlabel)+vidformat
        extract_subclip(video, begintime, endtime, targetname)
    if os.path.exists(video) == False:
        print('video ' + ID + 'does not exist!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!')
 
```
